```python
from scipy.ndimage import affine_transform
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.load('../data/carseq.npy')
img = data[:,:,0]
plt.imshow(img)
M = np.array([[1.0, 0.0, 1.0], [0.0, 1.0, 1.0]])
height, width = img.shape

```

```python
image1_warped = affine_transform(img, M, (width, height))
plt.imshow(image1_warped)
```

```python
image2_warped = cv2.warpAffine(img, M, (width, height))
plt.imshow(image2_warped)
```
